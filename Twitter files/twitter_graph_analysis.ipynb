```python
G_blm=nx.read_edgelist('./blm_1000/blacklivesmatter_graph_edgelist.csv',delimiter=',')

Gcc = sorted(nx.connected_components(G_blm), key=len, reverse=True)
G0 = G_blm.subgraph(Gcc[0])
G_blm = G0
Name = "#BlackLivesMatter"
print(Name)
Node_Count = G_blm.number_of_nodes()
Edge_Count = G_blm.number_of_edges()
print("Node Count=",Node_Count)
print("Edge Count=",Edge_Count)

sum1=0
deg=G_blm.degree()
for key, val in deg:
    sum1=sum1+val
Avg_deg = sum1/G_blm.number_of_nodes()
print("Average Degree=",Avg_deg)

t = nx.triangles(G_blm)
sum2=0
for i in t.values():
    sum2=sum2+i
Tri=int(sum2/3)
print("Number of Triangles=",Tri)
Num_Comp = nx.number_connected_components(G_blm)
print("Number of components=",nx.number_connected_components(G_blm))

Gcc = sorted(nx.connected_components(G_blm), key=len, reverse=True)
G0 = G_blm.subgraph(Gcc[0])
Lar_Comp=len(G0.nodes)
print("Size of largest connected components=",Lar_Comp)
Clus_Coeff=nx.average_clustering(G_blm)
print("Clustering Coefficient",Clus_Coeff)


den=nx.density(G_blm)
print("density",den)
ass=nx.degree_assortativity_coefficient(G_blm)
print("Assortativity",ass)

print("Degree Centrality: ")
deg_cen_dict={}
deg_cen_avg=0
sum3=0
deg_cen= nx.degree_centrality(G_blm)
c=0
for i in sorted(deg_cen, key=deg_cen.get,reverse=True):
    print((i,deg_cen[i]))
    deg_cen_dict[i]=(deg_cen[i])
    sum3=sum3+deg_cen[i]
    c=c+1
    if c>=5:
        break
deg_cen_avg=sum3/Node_Count

print("Betweenness Centrality: ")
bet_cen_dict={}
bet_cen_avg=0
sum4=0
bet_cen= nx.betweenness_centrality(G_blm)
c=0
for i in sorted(bet_cen, key=bet_cen.get,reverse=True):
    print((i,bet_cen[i]))
    bet_cen_dict[i]=(bet_cen[i])
    sum4=sum4+bet_cen[i]
    c=c+1
    if c>=5:
        break
bet_cen_avg=sum4/Node_Count

print("Closeness Centrality: ")
clo_cen_dict={}
clo_cen_avg=0
sum5=0
clo_cen= nx.closeness_centrality(G_blm)
c=0
for i in sorted(clo_cen, key=clo_cen.get,reverse=True):
    print((i,clo_cen[i]))
    clo_cen_dict[i]=(clo_cen[i])
    sum5=sum5+clo_cen[i]
    c=c+1
    if c>=5:
        break
clo_cen_avg=sum5/Node_Count

        
print("Eigenvector Centrality: ")
eig_cen_dict={}
eig_cen_avg=0
sum6=0
eig_cen= nx.eigenvector_centrality(G_blm)
c=0
for i in sorted(eig_cen, key=eig_cen.get,reverse=True):
    print((i,eig_cen[i]))
    eig_cen_dict[i]=(eig_cen[i])
    sum6=sum6+eig_cen[i]
    c=c+1
    if c>=5:
        break
eig_cen_avg=sum6/Node_Count

        
ith_el=[Name, Node_Count, Edge_Count, Avg_deg, Tri, Num_Comp, Lar_Comp, Clus_Coeff, den, ass, deg_cen_dict, deg_cen_avg, bet_cen_dict, bet_cen_avg, clo_cen_dict, clo_cen_avg, eig_cen_dict, eig_cen_avg]
db.loc[len(db)] = ith_el


```
